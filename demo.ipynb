# Bitwise operators

- `&` → sets bit to 1 if both bits are 1
- `|` → sets bit to 1 if either bit is 1
- `^` → sets bit to 1 if bits are different
- `~` → inverts all bits
- `<<` → shifts bits left (multiplies by 2)
- `>>` → shifts bits right (divides by 2)

````python
a = 5   # 0101 in binary
b = 3   # 0011 in binary

# Bitwise AND
print(a & b)   # 1  (0001)

# Bitwise OR
print(a | b)   # 7  (0111)

# Bitwise XOR
print(a ^ b)   # 6  (0110)

# Bitwise NOT
print(~a)      # -6 (inverts all bits)

# Left shift
print(a << 1)  # 10 (1010)

# Right shift
print(a >> 1)  # 2  (0010)
````

# Try-Except (Exception Handling)
**Description:**
- Used to handle runtime errors gracefully and prevent program from crashing.
- `try` block contains code that might raise an exception.
- `except` block contains code to handle the exception.
- `else` block executes if no exception occurs. *(Optional)*
- `finally` block executes regardless of exception. *(Optional)*

**Syntax:**
```python
try:
    # code that may raise an exception
except ExceptionType:
    # code to handle the exception
else:  # optional
    # executes if no exception occurs
finally:  # optional
    # executes no matter what


````python
# devision

def divide(a, b):
    return a / b


try:
    result = divide(10, 2)
    print("Division is:", result)
except ZeroDivisionError:
    print("Error: something went wrong during division.")
except TypeError:
    print("Error: Invalid input type for division.")
except:
    print("An unexpected error occurred.")
finally:
    print("Execution completed.")
````

# File Operations

**Description:**
- Python provides built-in functions to create, read, write, and manipulate files.  
- Files should generally be closed after operations, or you can use `with` to handle it automatically.

**Old Syntax:**
```python
# Open a file
file = open("filename.txt", "mode")

# Perform operations (read/write)

# Close the file
file.close()

```

**Modern Syntax:**
```python
with open("filename.txt", "mode") as file:
    # perform file operations
    file.write("Some text")  # for writing
    content = file.read()    # for reading
```
**Notes:**
- with modern syntax, no need to call `file.close()`; `with` handles it automatically.

- Works with all file modes (r, w, a, rb, wb, etc.).

### File Modes

| Mode  | Description                                                                 |
|-------|-----------------------------------------------------------------------------|
| `r`   | Read mode (default). Opens the file for reading. File must exist.           |
| `w`   | Write mode. Creates a new file or overwrites if it exists.                  |
| `a`   | Append mode. Adds data to the end of the file if it exists.                 |
| `rb`  | Read binary mode. Reads file in binary format.                              |
| `wb`  | Write binary mode. Writes file in binary format.                            |
| `r+`  | Read and write mode. File must exist.                                       |
| `w+`  | Write and read mode. Creates a new file or overwrites if it exists.         |
| `a+`  | Append and read mode. Opens file for reading and appending.                 |
| `ab`  | Append binary mode. Adds binary data to the end of the file.                |



````python
file = open("image.png", "rb")

file2 = open("new_image.png", "wb")

file2.write(file.read())
file2.close()
file.close()

````

````python
with open("image.png", "rb") as file:
    with open("copy_image.png", "wb") as file2:
        file2.write(file.read())



print("File copy completed.")
````

# HTTP Requests in Python (Using `requests` Library)

**Description:**  
- The `requests` library allows you to send HTTP requests in Python easily.  
- Common request types: `GET`, `POST`, `PUT`, `DELETE`.  
- You can send parameters, headers, and data, and handle responses.

**Installation:**  
```bash
pip install request
```

**Example**
```python
# GET request
response = requests.get("https://api.example.com/data")
print(response.status_code)  # HTTP status code
print(response.text)         # Response content as text
print(response.json())       # Parse response as JSON (if JSON)

# GET request with parameters
params = {"q": "python", "page": 1}
response = requests.get("https://api.example.com/search", params=params)

# POST request with data
data = {"username": "user1", "password": "pass123"}
response = requests.post("https://api.example.com/login", data=data)

# Custom headers
headers = {"Authorization": "Bearer YOUR_TOKEN"}
response = requests.get("https://api.example.com/protected", headers=headers)
```

**Important Notes:**

- `response.status_code` gives the HTTP status.

- `response.json()` parses JSON responses.

- `response.text` gives raw response content.

Always handle exceptions for network errors using `try-except`.

````python
import time
import requests

with open("github_users.txt", "r") as file:
    for user in file:
        user = user.strip()  # remove newline
        page = 1
        while True:
            url = f"https://api.github.com/users/{user}/repos?per_page=100&page={page}"
            response = requests.get(url)
            
            if response.status_code != 200:
                print(f"Error fetching repos for {user}: {response.status_code}")
                break

            data = response.json()
            if not data:  # empty list means no more pages
                break

            for repo in data:
                if isinstance(repo, dict):
                    print(repo.get("name"))

            page += 1
            time.sleep(0.1)
````
