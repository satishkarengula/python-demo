# Logical Operators

```python
age = 10

occupation = "student"

is_student = True

if age > 18 and occupation == "engineer":
    print("You are eligible for the professional engineering association.")

if age < 18 or occupation == "engineer":
    print("You are either a minor or an engineer.")

if not is_student:
    print("You are not a student.")
else:
    print("You are a student.")
```

# While Loop


```python
a = 5

while a < 10:
    print("a is:", a)
    a = a + 1
else:
    print("Loop ended. Final value of a is:", a)


print("End of the program.")
```

# For Loop

```python
for i in range(10):
    print("Iteration:", i)
    if i == 5:
        break
```
